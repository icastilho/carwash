# MODELO

## limpa descrição do modelo

```python
#Conta quantidade de versões 0 km
data.query('anomod == "32000"').shape
```

```python

dfmodelo = pd.DataFrame(data={'id':data['modelo_id'], 'desc':data['modelo'], 'year': data['anomod'], 'brand': data['marca'], 'value':data['valor'] })
```

```python
dfmodelo.head(5)
```

```python
#limpa descrição
dfmodelo_clean = dfmodelo.join(dfmodelo['desc'].str.extract('(?P<model>^[^\W]+)(?P<version>.*)', expand=True))
```

```python
## Verifica se algum resultado é nulo
for field in dfmodelo_clean.columns:
    print(field, 'NaN:',dfmodelo_clean[field].isnull().sum())
```

```python
dfmodelo_clean.loc[dfmodelo_clean['model'].isnull()]
```

```python
# Executa regex para modelos que não foram detectados por nao ter a versão 
#substitui os nulos pela descrição inicial
dfmodelo_clean['model'].fillna(dfmodelo_clean['desc'].str.extract('(?P<model>^[^\W]+)'), inplace=True)
```

```python
dfmodelo_clean['model'] = dfmodelo_clean['model'].str.title()
dfmodelo_clean
```

```python
## Verifica se algum resultado é nulo no resultado final
for field in dfmodelo_clean.columns:
    print(field, 'NaN:',dfmodelo_clean[field].isnull().sum())
```

```python
#remove desc que não será necessário para versão
dfversao = dfmodelo_clean.drop('desc', axis=1)
dfversao.to_csv('data/versions.csv', index=False)
```
